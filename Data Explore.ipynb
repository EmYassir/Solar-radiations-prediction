```python
import random

fig=plt.figure(figsize=(12, 12))
columns = 2; rows = 5
ind = 0
for i in range(1, columns):
    r = random.randint(0,len(dataset))
    for channel in [1,2,3,4,6]:
        ind+=1
        explore = dataset.iloc[r][['ncdf_path','hdf5_16bit_path','hdf5_16bit_offset']]
        hdf5_path = explore['hdf5_16bit_path']
        hdf5_offset = explore['hdf5_16bit_offset']  # this would correspond to: 2010.06.01.0800 + (32)*15min = 2010.06.01.1600
        with h5py.File(hdf5_path, "r") as h5_data:
          ch1_data = fetch_hdf5_sample("ch"+str(channel), h5_data, hdf5_offset)
        fig.add_subplot(rows, columns, ind)
        try:
            plt.title(hdf5_path.split("/")[-1] + " - " + str(hdf5_offset) + " - " + str(channel))
            plt.xticks([])
            plt.yticks([])
            plt.imshow(ch1_data, origin='lower',cmap='bone')
        except:
            print("maybe corrupted: ", hdf5_path + " - " + str(hdf5_offset))
plt.show()
```
